```python
import numpy as np
import matplotlib.pyplot as plt

# 실제 계산된 연간 사고 확률 기반 평균값 (λ)
lambda_per_year = 12 / (154 * 3)  # 약 0.0216

# 시뮬레이션: 1 API가 1년 동안 몇 번 사고 나는지 10,000번 시도
simulations = 10000
incident_counts = np.random.poisson(lam=lambda_per_year, size=simulations)

# 실제 사고 확률 (사고가 1건 이상일 확률)
real_prob = np.mean(incident_counts >= 1)
print(f"실제 사고 발생 확률 (1년 내 1건 이상):{real_prob * 100:.2f}%")

# 히스토그램 시각화
plt.hist(incident_counts, bins=range(0, 4), align='left', rwidth=0.5)
plt.title("API 1개당 연간 사고 발생 건수 시뮬레이션")
plt.xlabel("사고 건수")
plt.ylabel("빈도수 (시뮬레이션 수)")
plt.show()
```
